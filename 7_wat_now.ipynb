There are various algorithms for machine learning such as logistic regression, support vector machine, k-nearest neighbors, k-means clustering and more. These algorithms use iterations (or epochs) to improve the accuracy (or as you will learn, reducing the loss).

You'll need to know about actuators and sensors. So actuators are basically motors. You have DC motors, servo motors, stepper motors, brushless motors and some of them require additional components in order to run. For example, if you're making a simple car and you're gonna be using DC motors and an arduino (You really dont need an RPi for that, trust me, I learnt it the hard way) then you will also need a motor driver such as an L293d motor driver. Servos on the other hand dont need any drivers, however there may be a time when you need more than 10 servos for your project (maybe you're making a hexapod). For projects like that, your arduino might not have enough PWM pins (pins that send servos the signal to move to a position) in which case you'll have to use a 16 bit servo driver. So there are things like this you need to think about. 

Sensors are modules (not libraries) that send data to your microcontroller (RPi or arduino). They can be used as a way to run your robot car or hexapod or find their X, Y, Z positions. Even your phone can be used as a giant sensor bundle. That thing has everything starting from cameras, accelerometers, gyroscopes, magnetometer (fancy way of saying compass), GPS, touch sensors and so much more.
